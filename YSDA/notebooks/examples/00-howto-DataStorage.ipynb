# DataStorage

Here we want tell you about DataStorage class, how you can do different manipulations with data to use them in machine learning. Now you can use any type of data, only define converter and use data in analysis.

DataStorage is class which describes the following data modifications:

* `get_train_test(self, train_size=None, test_seize=None, random_state=13, group_columns=None)` - randomly split your data into train and test datasets using pseudo generator with `seed(random_state)`. Also elements from the same group (group by `group_columns`) will be in the same dataset


* `get_folds(self, ids, nfolds, random_state=13, group_columns=None)` - split data into `nfolds` folds and union `ind` (set of folds indices) using pseudo generator with `seed(random_state)`.  Also elements from the same group (group by `group_columns`) will be in the same dataset


* `static union(ds_tuple)` - union different data (`ds_tuple` - tuple of DataStorages)  into one


* `add_columns(self, data)` - add new columns to data storage (`data` - dict with key-name of new column and value-list of val


* `select(self, selection, str_type=PYTHON)` - do preselection for data


* `def shuffle(self, random_state=None)` - randomly shuffle your data


To get data in DataStorage you can use 

` get_data(features=None) ` - return pandas.DataFrame data with columns=`features`. 

Features can be existing names of branches, or any modification: `'new_variable: a / sin(b)'` - then return feature `new_variable` with values `a[i] / sin(b[i])`

To get columns of data also can use `[]: ds['feature'], ds[['feature1', 'feature2']]`

__NOTE: all operations don't modify ds and return new ds__

# Labeled DataStorage

For machine learning purposes you can use __LabeledDataStorage__.

* `__init__(self, ds, labels=None, sample_weight=None, random_state=13, shuffle=True)` - if `shuffle` is Ture then `random_state` is used to shuffle all data


* `def add(self, ds, labels=None, sample_weight=None)` - add data with special labels and weight. `labels` - 0 (bck) or 1 (signal), or array of labels, or name of column in data storage. `sample_weight` - float number, or array of weights, or name of column in data storage. 


* `def pop(self, index=-1)` - remove 


* `def get_data(self, features=None)` - return data (pandas.DataFrame)


* `def get_labels(self)` - return labels (numpy.ndarray)


* `def get_weights(self)` - return weights (numpy.ndarray)


```python
from rep.data_storage import LabeledDataStorage
dataset = LabeledDataStorage(train_bck, 0, 0.1)
dataset.add(train, 1, 1.1)
dataset.get_data(['FL: FlightDistanceError*2', 'DOCAone'])
```

```python
dataset.get_labels()
```

```python
dataset.get_weights()
```

# Data Information

`DataInfo` - class, which allows you to see on features distribution, their correlation in your labeled dataset

* `def features_pdf(self, features, bins=30) :return plotting.GridPlot(plotting.ErrorPlot):`
        
* `def features_correlation_matrix(self, features, vmin=-1, vmax=1) :return plotting.GridPlot(plotting.ColorMap):`
        
* `def scatter(self, correlation_pairs, size=20, alpha=0.1) :return plotting.GridPlot(plotting.ScatterPlot)`
        
Using this function you get `plotting` object which has method plot() to present pictures

```python
# import this class
from rep.data_report import DataInfo

data_report = DataInfo(dataset)
```

### Features distributions

```python
data_report.features_pdf(["p0_pt", "p1_pt"]).plot()
```

### Features correlation matrix

```python
data_report.features_correlation_matrix(["p0_pt", "p1_pt", "p2_pt", "CDF3", "CDF1", "CDF2"]).plot(figsize=(14, 4), show_legend=False)
```

```python
data_report.scatter([('CDF1', 'CDF2'), ('CDF1', 'CDF3')], alpha=0.01).plot(figsize=(14, 5))
```
